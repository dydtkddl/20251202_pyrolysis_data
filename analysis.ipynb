```python
import pandas as pd
import logging
from tqdm import tqdm

logging.basicConfig(level=logging.INFO, format="%(asctime)s | %(levelname)s | %(message)s")

# ----------------------------------------------------------
# 1) GRADE 컬럼 이름 정리
# ----------------------------------------------------------
df = pd.read_excel("./20251204.subset_with_groups20251202-검수.xlsx")

DF = df.rename(columns={"Unnamed: 50": "GRADE"})

# ----------------------------------------------------------
# 2) GRADE NaN 제거
# ----------------------------------------------------------
DF_clean = DF.dropna(subset=["GRADE"])

logging.info(f"Original rows: {len(DF)}")
logging.info(f"Clean rows (GRADE not null): {len(DF_clean)}")

# ----------------------------------------------------------
# 3) GRADE 통계
# ----------------------------------------------------------
grade_counts = DF_clean["GRADE"].value_counts()
grade_ratio = DF_clean["GRADE"].value_counts(normalize=True).round(3)

logging.info("\n=== GRADE Counts ===\n" + str(grade_counts))
logging.info("\n=== GRADE Ratio ===\n" + str(grade_ratio))

# ----------------------------------------------------------
# 4) GROUP + GRADE 교차표
# ----------------------------------------------------------
cross = pd.crosstab(DF_clean["GROUP"], DF_clean["GRADE"])

logging.info("\n=== GROUP × GRADE ===\n" + str(cross))

# ----------------------------------------------------------
# 5) 저장 (선택)
# ----------------------------------------------------------
cross.to_csv("group_grade_stats.csv", encoding="utf-8-sig")
```
